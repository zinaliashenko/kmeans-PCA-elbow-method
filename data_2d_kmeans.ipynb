```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data_2d = pd.read_csv('data_2d.csv', header=None)
```

```python
data_2d.head()
```

```python
data_2d.shape
```

```python
data_2d[0].unique()
```

```python
n_clusters = list(range(1,10))
distortians = []

for i in n_clusters:
    km = KMeans(n_clusters=i, init='k-means++', max_iter=100, random_state=42, n_init='auto')
    km.fit(data_2d)
    distortians.append(km.inertia_)
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot()
ax.plot(n_clusters, distortians, marker='o')
plt.title('Elbow Method for "data_2d.csv"')
plt.xlabel('Number of Clusters')
plt.ylabel('Distortians')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2, random_state=42, n_init='auto')
kmeans.fit(data_2d)

print(kmeans.labels_[0:10])
print(kmeans.inertia_)
print(kmeans.cluster_centers_)
```

```python
plt.scatter(data_2d[1], data_2d[2], alpha=0.7)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='r', marker='X', s=200)
plt.title('K-means "data_2d.csv"')
plt.show()
```
